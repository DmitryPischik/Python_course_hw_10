```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
unique_values = data['whoAmI'].unique()
unique_values
```

```python
for value in unique_values:
    data[f'I_am_{value}'] = data['whoAmI'].apply(lambda x: 1 if x == value else 0)
data.head()
```

```python
data.drop(data.columns[[len(unique_values)]], axis=1, inplace=True)
data.head()
```
